```python
import matplotlib.pyplot as plt
import numpy as np
import math
#Grace Horton
#Homework 4
#Collaborators: None

file = open("mat5.txt")
A = np.loadtxt(file, delimiter=",")
size = np.size(A)
rows = len(A)
cols = len(A[0])
nnz = np.count_nonzero(A)
N = rows

print('************* HW 4 *****************')
print('Q1:')
print('Matrix Size:', size)
print('Matrix Dimensions:', rows,'x',cols)
print('====================================')
print('Q2:')
print('Number of Non-zeros:', nnz)
print('====================================')
print('Q3:')
print('Is the Matrix symmetric? Yes/No')
print('..........')

def isOrthogonal(a, m, n) :
    if (m != n) :
        return False
     
    trans = [[0 for x in range(n)]
                for y in range(n)]
                 
    #transpose
    for i in range(0, n) :
        for j in range(0, n) :
            trans[i][j] = a[j][i]
             
    prod = [[0 for x in range(n)]
               for y in range(n)]
                
    # product of a[][] and its transpose 
    for i in range(0, n) :
        for j in range(0, n) :
     
            sum = 0
            for k in range(0, n) :
                sum = sum + (a[i][k] * a[j][k])
     
            prod[i][j] = sum

    # identity matrix check
    for i in range(0, n) :
        for j in range(0, n) :
 
            if (i != j and prod[i][j] != 0) :
                return False
            if (i == j and prod[i][j] != 1) :
                return False
 
    return True

#function for finding the transpose, for convenience 
def transpose(mat, tr, N):
	for i in range(N):
		for j in range(N):
			tr[i][j] = mat[j][i]

# Returns true if given Matrix is symmetric, else false
def isSymmetric(mat, N):
	
	tr = [ [0 for j in range(len(mat[0])) ] for i in range(len(mat)) ]
	transpose(mat, tr, N)
	for i in range(N):
		for j in range(N):
			if (mat[i][j] != tr[i][j]):
				return False
	return True

def isDiagonalMatrix(mat) :
    for i in range(0, N):
        for j in range(0, N) :
            # checks if elements outside main diagonal are zero or not.
            if ((i != j) and
             (mat[i][j] != 0)) :
                return False
 
    return True

if (isSymmetric(A, N)):
	print("Yes")
else:
	print("No")

print('====================================')
print('Q4:')
print('Is the Matrix Diagonal? Yes/No')
print('..........')

if (isDiagonalMatrix(A)):
	print("Yes")
else:
	print("No")

print('====================================')
print('Q5:')
print('Is the Matrix Orthogonal? Yes/No')
print('..........')

if (isOrthogonal(A,rows,cols)):
	print("Yes")
else:
	print("No")

print('====================================')
print('Q6:')
# Singular-value decomposition
U, s, VT = np.linalg.svd(A)
rank = np.linalg.matrix_rank(A)
#rank = len(s)
smol = min(s)
big = max(s)
cond = big/smol

print('Rank:', rank)
print('====================================')
print('Q7:')
print('Smallest Singular Value:', smol)
print('====================================')
print('Q8:')
print('Largest Singular Value:', big)
print('====================================')
print('Q9:')
print('Condition Number:', cond)
print('====================================')
print('Q10:')

b1 = np.random.randint(100,size=(rows,1))
b2 = np.random.randint(100,size=(rows,1))
b3 = np.random.randint(100,size=(rows,1))
b4 = np.random.randint(100,size=(rows,1))
b5 = np.random.randint(100,size=(rows,1))
x1 = np.linalg.solve(A,b1)
x2 = np.linalg.solve(A,b2)
x3 = np.linalg.solve(A,b3)
x4 = np.linalg.solve(A,b4)
x5 = np.linalg.solve(A,b5)
check1 = np.allclose(np.dot(A,x1),b1)
check2 = np.allclose(np.dot(A,x2),b2)
check3 = np.allclose(np.dot(A,x3),b3)
check4 = np.allclose(np.dot(A,x4),b4)
check5 = np.allclose(np.dot(A,x5),b5)
print('x1 Solution Check:', check1)
print('x2 Solution Check:', check2)
print('x3 Solution Check:', check3)
print('x4 Solution Check:', check4)
print('x5 Solution Check:', check5)
print('====================================')

xval = A[np.matrix.nonzero(A)]
#print('x values:', xval)
print('Solutions::')
#print('x1:', x1)
#print('x2:', x2)
#print('x3:', x3)
#print('x4:', x4)
print('x5:', x5)
print('====================================')  
print('************ Matrix Plots ************')
print('====================================')

#plt.spy(A)
plt.subplot(2,1,1)
plt.plot(xval,"*")
plt.title("Non-Zero Elements")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()

print('====================================')

plt.subplot(2,1,2)
plt.plot(abs(A),"*")
plt.title("Magnitude of Elements")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()

```
